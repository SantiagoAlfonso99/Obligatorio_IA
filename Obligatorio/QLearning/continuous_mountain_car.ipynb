Esta notebook contiene bloques de código útiles para realizar Q-learning en el entorno "Continuous Mountain Car"

```python
import numpy as np
from continuous_mountain_car_env_extended import ContinuousMountainCarEnvExtended
import random 
import importlib
import q_learning_agent
importlib.reload(q_learning_agent)
from q_learning_agent import QLearningAgent
import pickle
```

```python
# Cambiar render_mode a rgb_array para entrenar/testear
env = ContinuousMountainCarEnvExtended(render_mode='rgb_array')
```

Observation Space

```python
env.observation_space
```

Action Space

```python
env.action_space
```

```python
x_space = np.linspace(-1.2, 0.6, 10)
vel_space = np.linspace(-0.07, 0.07, 10)
actions = list(np.linspace(-1, 1, 3))
qLearningAgent = QLearningAgent(x_space, vel_space, actions)
```

```python
rewards = qLearningAgent.train_agent(env, 10, .9, .9, .9)
with open('modelo_entrenado.pkl', 'wb') as archivo:
    pickle.dump(qLearningAgent, archivo)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(range(len(rewards)), rewards, label="Total Reward per Episode")
plt.xlabel("Episodes")
plt.ylabel("Total Reward")
plt.title("Learning Curve for Q-Learning Agent")
plt.legend()
plt.grid()
plt.show()
```
